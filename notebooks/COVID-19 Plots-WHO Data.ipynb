```python
# Similar to the other notebook but the data is sourced directly from WHO.  See ../who/README.md
import os
import pandas as pd
```

```python
# Load Data
datapath = '../who/total_cases.csv'
    
df = pd.read_csv(datapath)
#last_date = list(df.columns)[-1]
print(df['Canada'].iloc[-1])
print(df.columns)
df.index=df['date']
del df['Worldwide']
del df['International']

selected = [c for c in df.columns if c!='date' and df[c].iloc[-1]>100]

print(selected)
filtered = df[selected]

df=filtered
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [10, 10]
```

```python

for c in df.columns:    
    for i,a in enumerate(df[c]):        
        #if i==0 and a>100:
        #    df[c]=df[c].shift(4) # hack to push out china
        #    break
        if a>=100:
            df[c]=df[c].shift(-i)            
            break
df=df.reset_index(drop=True)
df
```

```python
from math import ceil
fig, ax = plt.subplots()
markers = [item for sublist in ['oxds']*ceil(len(df.columns)/4) for item in sublist]
#df.plot(logy=True, ax=ax) #, ax=ax)
for c,m in zip(df,markers):
    df[c].plot(logy=True, marker=m,  ax=ax)

ax.legend()
```
